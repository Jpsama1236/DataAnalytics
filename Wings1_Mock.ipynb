```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/bank (1).csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df1=df[(df['month']=='apr')|(df['month']=='may')|(df['month']=='jun')]
plot1=sns.lineplot(x='campaign',y='day',data=df1,hue='month',marker='o')
plot1.set_title('Campaign Vs Duration - Month Wise',size=15)
plot1.set_xlabel('Campaign',size=12)
plot1.set_ylabel('Duration',size=12)
```

```python
df3=pd.DataFrame()
df3['h_yes']=df[df['housing']=='yes']['job'].value_counts()
df3['h_no']=df[df['housing']=='no']['job'].value_counts()
df3['l_yes']=df[df['loan']=='yes']['job'].value_counts()
df3['l_no']=df[df['loan']=='no']['job'].value_counts()
plot3=df3.plot.bar()
df3
```

```python
#fig,ax=plt.subplots(figsize(10,10))
df4=df[(df['month']=='jan')|(df['month']=='feb')|(df['month']=='mar')|(df['month']=='apr')|(df['month']=='may')]
plot4=sns.scatterplot(x='campaign',y='duration',data=df4,hue='month',hue_order=['jan','feb','mar','apr','may'])
plot4.set_title('Campaign Vs Duration - Month Wise',size=15)
plot4.set_xlabel('Campaign',size=12)
plot4.set_ylabel('Duration',size=12)
```

```python
plot5=sns.lmplot(x='balance',y='age',data=df,hue='deposit')
```

```python
df6=df[(df['job']=='blue-collar')|(df['job']=='retired')|(df['job']=='technician')|(df['job']=='admin.')|(df['job']=='management')]
sns.histplot(data = df6,x = 'duration',hue='job',binwidth = 100)
```

```python
df7=df[(df['job']=='admin.')]
sns.histplot(data = df7,x = 'duration')
```
